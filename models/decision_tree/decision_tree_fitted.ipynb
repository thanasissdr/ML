```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

# Load data

```python
data = load_iris(as_frame=True)

X = data["data"]
y = data["target"]
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 8))

plot_tree(
    clf, feature_names=data.feature_names, class_names=data.target_names, filled=True
)
```

# How to interpret the tree structure

$$\Pr(y=\text{setosa}\mid \text{petal length (cm) $\le$ 2.75}) = \frac{50}{50} = 1$$

```python
mask = X["petal length (cm)"] <= 2.75
display(y[mask].value_counts())

display(y[~mask].value_counts())
```

$$\Pr(y=\text{virginica}\mid \text{petal length (cm) $>$2.75 & petal width (cm) $>$ 1.75}) = \frac{45}{46} = 0.9782$$

```python
mask = (X["petal length (cm)"] > 2.75) & (X["petal width (cm)"] > 1.75)
y[mask].value_counts()
```
